```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/ml_1data.csv")
df
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
%matplotlib inline
plt.xlabel('area(in sqft)')
plt.ylabel('price(in Rs)')
plt.scatter(df.area,df.price,color= 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[4638]])
```

```python
area_to_predict = np.array([[4638]])
predicted_price = reg.predict(area_to_predict)
print(predicted_price)
```

```python
%matplotlib inline
plt.xlabel('area(in sqft)')
plt.ylabel('price(in Rs)')
plt.scatter(df.area,df.price,color= 'red', marker = '+')
plt.plot(df.area,reg.predict(df[['area']]), color= 'blue')
plt.show()
```

```python
df2 = pd.read_csv('/content/ml_2data.csv')
df2
```

```python
p = reg.predict(df2)
df2['price'] = p
df2
```

```python
df2.to_csv("prediction.csv")
```
