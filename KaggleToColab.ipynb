```python
import os
import random
import SimpleITK as sitk

# Define actual paths
input_dir = '/content/drive/MyDrive/kaggle/no'  # Replace with your input directory path
output_dir = '/content/drive/MyDrive/kaggle/Truncated'  # Replace with your output directory path

# List all files in the input directory
file_list = os.listdir(input_dir)

# Choose 100 random files from the file_list
random_files = random.sample(file_list, 25)

# Iterate through each randomly selected file (assuming they are all MRI images)
for file_name in random_files:
    if file_name.endswith('.nii.gz'):  # Adjust based on your file format
        # Load the image
        full_path = os.path.join(input_dir, file_name)
        image = sitk.ReadImage(full_path)

        # Convert image to numpy array if needed
        image_array = sitk.GetArrayFromImage(image)

        # Example preprocessing steps (resizing and normalization)
        resized_image = sitk.GetImageFromArray(image_array)
        new_size = (128, 128, 64)  # Adjust the dimensions as needed
        resized_image.SetSpacing((image.GetSpacing()[0] * image.GetWidth() / new_size[0],
                                  image.GetSpacing()[1] * image.GetHeight() / new_size[1],
                                  image.GetSpacing()[2] * image.GetDepth() / new_size[2]))
        resized_image.SetDirection(image.GetDirection())

        # Save the processed image
        output_path = os.path.join(output_dir, file_name)
        sitk.WriteImage(resized_image, output_path)

        print(f"Processed and saved: {file_name}")

print("Processing completed.")
```
